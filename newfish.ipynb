```python
db = client.fishpedia
collection = db.fish_freshwater
data = pd.DataFrame(list(collection.find({}, {'_id': False})))
```

```python
data.head()
```

```python
#remove the "(" in the name
data['name']=data['name'].str.replace("(","", regex= True)
data.head()
```

```python
#remove "\n" in dataframe
data=data.replace("\n","", regex= True)
# create new columns with american mesures
data['size inches']= data["size"].str.extract(r'\((.*?)\)')
data['min. tank size in US Gallon']= data["min. Tank size"].str.extract(r'\((.*?)\)')
data['temperature in °F']= data["temperature"].str.extract(r'\((.*?)\)')
# delete the american mesures in the origin colums
data['size']=data['size'].str.replace(r'\((.*?)\)', "")
data['min. Tank size']=data['min. Tank size'].str.replace(r'\((.*?)\)', "")
data['temperature']=data['temperature'].str.replace(r'\((.*?)\)', "", regex= True)
# extract the numeric of the tank size
data['min. Tank size'] = data['min. Tank size'].str.extract(r'(\d+)', expand=False)
data['min. tank size in US Gallon'] = data['min. tank size in US Gallon'].str.extract(r'(\d+)', expand=False)
```

```python
data.head()
```

```python
data['size min in cm']= data["size"].str.extract(r'(.*)-')
data['size max in cm']= data["size"].str.extract(r'-(.*)')
data['size max in cm']= data['size max in cm'].replace("cm","", regex= True)

data.head()
```

```python
data['PH min.']= data["PH"].str.extract(r'(.*)-')
data['PH max']= data["PH"].str.extract(r'-(.*)')
data.head()
```

```python
data['temperature Min in °C']= data["temperature"].str.extract(r'(.*)-')
data['temperature Max in °C']= data["temperature"].str.extract(r'-(.*)')
data['temperature Max in °C']= data['temperature Max in °C'].replace("°C","", regex= True)
data.head()
```

```python
data['water hardness Min ']= data["water hardness"].str.extract(r'(.*)-')
data['water hardness Max']= data["water hardness"].str.extract(r'-(.*)')
data['water hardness Max']= data['water hardness Max'].replace("°d","", regex= True)
data.head()
```

```python
data['life span Min in years']= data["life span"].str.extract(r'(.*)-')
data['life span Max in years']= data["life span"].str.extract(r'-(.*)')
data['life span Max in years']= data['life span Max in years'].replace("years","", regex= True)
data.head()
```

```python
#retirer les doublons
data.drop_duplicates(keep = 'first', inplace=True)
data.shape
```
